```python
import pandas as pd
df= pd.read_csv("homeprices.csv")
df
```

```python
# create dummy variable columns
dummies = pd.get_dummies(df.town)
dummies
```

```python
# concatenate dummies df with original dataframe
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# rule of thumb - drop one dummy variable column if one variable can be derived from the rest of the variables
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
# create a linear regression model
from sklearn.linear_model import LinearRegression

# create an object of LinearRegression class
model = LinearRegression()
```

```python
# give X and Y for training
# x - area, monroe township, robinsville (independent variables)
# y - price (since it's the dependent variable)

X = final.drop('price', axis='columns')
X
```

```python
Y = final.price
Y
```

```python
# training the ML model
model.fit(X,Y)
```

```python
# predict the price of a 2800 sqr ft home in robinsville
model.predict([[2800,0,1]])
```

```python
# predict the price of a 3400 sqr ft home in west windsor
model.predict([[3400,0,0]])
```

```python
# check how accurate your model is: model.score(x,y)
# compares the predicted value with the actual value
# 1.0000 means it's perfect
model.score(X,Y)
```
